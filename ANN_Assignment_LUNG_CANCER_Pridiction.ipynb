```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
```

```python
seed = 9
np.random.seed(seed)
```

```python
X = df.iloc[:,0:15]
Y = df.iloc[:,15]
```

```python
#Creating appropriate model based on the outcome needed
model = Sequential()
model.add(Dense(16, input_dim=15,  activation='relu')) #1st layer
model.add(Dense(12,  activation='relu')) #2nd layer
model.add(Dense(8, activation='relu')) #3rd layer
model.add(Dense(6, activation='relu')) #4th layer
model.add(Dense(3, activation='relu')) #5th layer
model.add(Dense(1, activation='sigmoid')) #6th layer
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X, Y, validation_split=0.30, epochs=120, batch_size=20)
```

```python
#Evaluating the model based on the accuracy %age
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
history.history.keys()
```

```python
#Summarizing history for the accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
#Summarizing history for the loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
